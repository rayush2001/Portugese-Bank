## 5.5 Scaling

```python
from sklearn.preprocessing import StandardScaler

# Storing encoded data in X variable
X = data

# Select only numeric columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform
X_scaled = X.copy()
X_scaled[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

```python
X_scaled
```

## 5.6 Encoding

```python
# One-Hot-Encoding using pandas
df_encoded = pd.get_dummies(X_scaled, columns=[
    'job', 'marital', 'contact', 'month',
    'day_of_week', 'poutcome', 'default', 'housing', 'loan', 'previous_contacted'
], drop_first=True, dtype=int)  # drop_first to avoid dummy variable trap
```

```python
education_map = {
    'illiterate': 0,
    'basic.4y': 1,
    'basic.6y': 2,
    'basic.9y': 3,
    'high.school': 4,
    'professional.course': 5,
    'university.degree': 6,
    'unknown': -1  # or use np.nan if you want to impute later
}

df_encoded['education_num'] = df_encoded['education'].map(education_map)
```

```python
df_encoded.drop(columns='education', inplace=True)
```

```python
# Encode target
df_encoded['y'] = df_encoded['y'].map({'no': 0, 'yes': 1})
```

```python
df_encoded.info()
```

```python
# Removing duration column because it can leak data
df_encoded.drop(columns='duration',inplace=True)
```

# 7. Data Splitting

## 7.1 X,y split

```python
X = df_encoded.drop('y',axis=1)
y = df_encoded['y']
```

## 7.2 train, test split

```python
# Importing train_test_split
from sklearn.model_selection import train_test_split

# It split 80% data for train data and 20% data for test data
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=14)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## 7.2 Imbalance Data 

```python
# Here we handling imbalance data
# Importing SMOTE from imblearn module
from imblearn.over_sampling import SMOTE

# Apply SMOTE on the training set only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
print(X_train_resampled.shape)
print(y_train_resampled.shape)
```

# 8. Model Implementation

## 8.1 Logistic Regression

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Initializing LogisticRegression model
LogReg =LogisticRegression(penalty = 'l2' ,C=0.5)

# Fitting the training data to the Logistic Regression model
LogReg.fit(X_train_resampled,y_train_resampled)

# predicting the target for testing data
y_pred=LogReg.predict(x_test)
```

```python
# Importing accuracy_score
from sklearn.metrics import accuracy_score

# Accuracy score for Logistic Regression
log_acc = accuracy_score(y_pred, y_test)
print(log_acc)
```

```python
# Importing classification_report
from sklearn.metrics import classification_report

# Classification report for logesticRegression
print(classification_report(y_test,y_pred))
```

# 10. Report on challenges faced

1. The dataset was highly imbalanced, with a significantly larger number of "no" responses compared to "yes" responses in the target variable. To  address this issue and avoid bias toward the majority class, I applied the SMOTE (Synthetic Minority Over-sampling Technique) method. This helped balance the class distribution and improved the model's ability to generalize without overfitting to the majority class.

2. Due to class imbalance, accuracy alone wasn’t reliable. I switched to F1 score, which better balances precision and recall for more meaningful evaluation.

# Conclusion

Gradient Boosting gave the best accuracy (87.92%) among all models, followed by Random Forest and Decision Tree. Naive Bayes performed the worst. Overall, ensemble models worked better than individual ones, showing they handle complex data more effectively. Gradient Boosting is the most reliable choice for this dataset.
